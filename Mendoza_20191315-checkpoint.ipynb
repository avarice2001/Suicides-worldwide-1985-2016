### Primeras impresiones
Noruega del año 1986 al 2014 (28 años) redujo su nivel de suicidios en 31 (en dicho periodo Noruega sextuplicó su PBI), mientras que Albania en 23 años aumento su nivel de suicidios en algo más de 21 (se debe tener en cuenta que Albania aproximadamente quintuplico su PBI). ### Jutando las 4 datas creadas
Usare el comando "merge" de la forma "outer" (ya que hare una union de dataframes) para juntar las 4 mini-bases de datos para poder realizar el gráfico pertinente. Lo hare en tres pasos y en cado uno usare el mismo comando y la misma logica.

### Junto a Noruega del año 1986 con Noruega del año 1986

```python
noruega = pd.merge(noruegos2014,
                     noruegos1986,
                    how = 'outer', indicator = True)
noruega
```

### Junto a Albania de 1987 con Albania al año 2010

```python
albania = pd.merge(albania1987,
                     albania2010,
                    how = 'outer', indicator = True)
albania
```

### Finalmente: Junto a Noruega y Albania para los 4 años

```python
paises = pd.merge(noruega,
                  albania,
                    how = 'outer', indicator = False)
paises
```

### Graficos
Usaremos el paquete "Altair" para hacer los graficos.

```python
import altair as alt
```

#### Grafico de dispersion
Para corroborar lo que dije más arriba (que no hay ninguna relación aparente entre el PBI per capita y los suicidios) uso el comando "mark_point()".

```python
alt.Chart(paises).mark_point(color='green').encode(
    x = alt.X("gdpc"),
    y = alt.Y("suicides"),
    color = alt.Color("country", 
            legend=alt.Legend(orient='bottom', titleOrient='left')),
    column = alt.Column("country")
    ).properties(width=300, height=300)
```

Se puede ver que el grafico no tiene ni pendiente negativa ni positiva por lo que se puede decir que no hay ninguna correlación entre el PBI per capita y los suicidios ("felicidad").

#### Grafico de barras
Para ver el numero de suicidios por sexo y país. Uso el comando "mark_bar()".

```python
alt.Chart(paises).mark_bar().encode(
    x = alt.X("sex"),
    y = alt.Y("suicides"),
    color = alt.Color("country", 
            legend=alt.Legend(orient='bottom', titleOrient='left')),
    column = alt.Column("country")
    ).properties(width=300, height=300)
```

Asimismo, a partir del grafico de barras se puede observar que el sexo con mas numeros de suicidios tanto en el "peor" (Albania) y "mejor" (Noruega) pais es el masculino.
